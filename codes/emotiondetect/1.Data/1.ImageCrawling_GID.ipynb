```python
from google_images_download import google_images_download as gid
```

```python
def download(keyword:str, limit:int = 100, format:str = "jpg") -> None:
    arguments = {
        "keywords":keyword,
        "limit":limit,
        "print_urls":True,
        "format":format,
        "output_directory":"../images",
    }
    print(gid.googleimagesdownload().download(arguments))
```

```python
def main():
    keywords:list = [
        "tranquil",
        "angry",
        "sad",
        "happy",
        "gloomy",
        "surprised",
        "fear",
        "confused",
        "tired",
    ]
    for nk in keywords:
        download(keyword = f"{nk} person")
```

```python
if __name__ == "__main__":
    main()
```
